```python
import csv
with open("sales.csv","r") as file:
    reader=csv.DictReader(file)
    salesData=(list(reader))
import csv
with open("product.csv","r") as file:
    reader=csv.DictReader(file)
    productData=(list(reader))
combined_data = []
total_sales=0
total_revenue=0
total_expenditure=0
profit=0
for sale in salesData:
    for product in productData:
        if sale['product_id'] == product['product_id']:
            quantity_sold=float(sale["quantity"])
            cost=float(product['cost_price'])
            price=float(sale["price"])
            total_revenue+=quantity_sold*price
            total_expenditure+=quantity_sold*cost
            profit+=total_revenue-total_expenditure
            combined_entry = {
                'sale_id': sale['sale_id'],
                'product_id': sale['product_id'],
                'quantity_sold': sale['quantity'],
                'sale_date': sale['date'],
                'category': product['category'],
                'product_name': product['product_name'],
                'cost' : product['cost_price'],
                'price': sale['price'],
                'total_revenue': total_revenue,
                'total_expenditure': total_expenditure,
                'profit' : profit
            }
            combined_data.append(combined_entry)
for entry in combined_data:
    print(entry)
import csv
with open("combined_data.csv","w",newline='') as file:
    writer=csv.DictWriter(file,["sale_id","product_id","quantity_sold","sale_date","category","product_name","cost","price","total_expenditure","total_revenue","profit"])
    writer.writeheader()
    writer.writerows(combined_data)
```
