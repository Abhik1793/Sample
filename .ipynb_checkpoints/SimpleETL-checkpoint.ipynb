```python
import csv
from datetime import datetime
from collections import defaultdict
with open("C:/Users/sales.csv","r") as file:
    reader=csv.DictReader(file)
    salesData=(list(reader))
import csv
with open("C:/Users/product.csv","r") as file:
    reader=csv.DictReader(file)
    productData=(list(reader))
combined_data = []
total_sales=0
total_revenue=0
total_expenditure=0
category_sales={}
category_profits = {}
product_sales = {}
customer_sales = {}
daily_sales = defaultdict(float)
monthly_sales = defaultdict(float)
quarterly_sales = defaultdict(float)
profit=0
def get_quarter(month):
    return f"Q{(month - 1) // 3 + 1}"
for sale in salesData:
    for product in productData:
        if sale['product_id'] == product['product_id']:
            quantity_sold=float(sale["quantity"])
            cost=float(product['cost_price'])
            price=float(sale["price"])
            category=product['category']
            customer = sale['customer_id']
            product_name = product['product_name']
            total_revenue+=quantity_sold*price
            total_expenditure+=quantity_sold*cost
            profit+=total_revenue-total_expenditure
            
            # Parse sale date
            sale_date = datetime.strptime(sale['date'], "%m/%d/%Y")
            day = sale_date.date()
            month = sale_date.strftime("%Y-%m")  # e.g., 2025-01
            quarter = f"{sale_date.year}-Q{(sale_date.month - 1) // 3 + 1}"  # e.g., 2025-Q1

            # Update sales dictionaries
            daily_sales[day] += total_revenue
            monthly_sales[month] += total_revenue
            quarterly_sales[quarter] += total_revenue
            
            #calculating category sales and updating
            if category in category_sales:
                category_sales[category]+=quantity_sold
                category_profits[category] += (quantity_sold * (price - cost))
            else:
                category_sales[category]=quantity_sold
                category_profits[category] = quantity_sold * (price - cost)
                
            #calculating product wise sales
            if product_name in product_sales:
                product_sales[product_name] += quantity_sold
            else:
                product_sales[product_name] = quantity_sold
                
            #calculating and updating customer sales:
            if customer in customer_sales:
                customer_sales[customer] += quantity_sold
            else:
                customer_sales[customer] = quantity_sold
                
            #Appending combined data
            combined_entry = {
                'sale_id': sale['sale_id'],
                'product_id': sale['product_id'],
                'quantity_sold': sale['quantity'],
                'sale_date': sale['date'],
                'category': product['category'],
                'product_name': product['product_name'],
                'cost' : product['cost_price'],
                'price': sale['price'],
                'total_revenue': total_revenue,
                'total_expenditure': total_expenditure,
                'profit' : profit
            }
            combined_data.append(combined_entry)

# Define a function to get top N items
def get_top_items(data_dict, top_n):
    top_items = []
    for _ in range(top_n):
        max_item = None
        max_value = 0
        for item, value in data_dict.items():
            if item not in [i for i, _ in top_items] and value > max_value:
                max_item = item
                max_value = value
        if max_item:
            top_items.append((max_item, max_value))
    return top_items

# Get the top 3 categories
top_categories = get_top_items(category_sales, 3)

# Get the top 3 products
top_products = get_top_items(product_sales, 3)

# Get the top 3 customers
top_customers = get_top_items(customer_sales, 3)

# Print results
print("\nTop 3 categories with highest sales:")
for category, sales in top_categories:
    print(f"Category: {category}, Total Sales: {sales}")

print("\nTotal Sales and Profit by Category:")
for category in category_sales:
    print(f"Category: {category}, Total Sales: {category_sales[category]}, Total Profit: {category_profits[category]:.2f}")

print("\nTop 3 products with highest sales:")
for product, sales in top_products:
    print(f"Product: {product}, Total Sales: {sales}")

print("\nTop 3 customers with highest purchases:")
for customer, sales in top_customers:
    print(f"Customer ID: {customer}, Total Purchases: {sales}")

# Print Sales Trends (Daily, Monthly, Quarterly)
print("\nDaily Sales Trends:")
for day, revenue in daily_sales.items():
    print(f"{day}: ${revenue:.2f}")

print("\nMonthly Sales Trends:")
for month, revenue in monthly_sales.items():
    print(f"{month}: ${revenue:.2f}")

print("\nQuarterly Sales Trends:")
for quarter, revenue in quarterly_sales.items():
    print(f"{quarter}: ${revenue:.2f}")

# printing the combined data
for entry in combined_data:
    print("\n",entry)

#Preparing summarized report
import csv
with open("combined_data.csv","w",newline='') as file:
    writer=csv.DictWriter(file,["sale_id","product_id","quantity_sold","sale_date","category","product_name","cost","price","total_expenditure","total_revenue","profit"])
    writer.writeheader()
    writer.writerows(combined_data)

# Initialize dictionaries to store sales data
daily_sales = defaultdict(float)
monthly_sales = defaultdict(float)
quarterly_sales = defaultdict(float)

for sale in combined_data:
    sale_date = datetime.strptime(sale['sale_date'], "%m/%d/%Y")
    
    # Extract day, month, and quarter
    day = sale_date.date()
    month = sale_date.strftime("%Y-%m")  # e.g., 2024-09 for September
    quarter = f"Q{(sale_date.month - 1) // 3 + 1}"
    
    # Update daily sales data
    daily_sales[day] += float(sale['total_revenue'])  # Adding total revenue for the day
    # Update monthly sales data
    monthly_sales[month] += float(sale['total_revenue'])  # Adding total revenue for the month
    # Update quarterly sales data
    quarterly_sales[quarter] += float(sale['total_revenue'])  # Adding total revenue for the quarter

summarized_data = []

for day, revenue in daily_sales.items():
    summarized_data.append({
        'period': day,
        'period_type': 'Daily',
        'total_revenue': revenue,
        'total_expenditure': sum([sale['total_expenditure'] for sale in combined_data if sale['sale_date'] == str(day)]),  # Aggregating expenditure per day
        'profit': sum([sale['profit'] for sale in combined_data if sale['sale_date'] == str(day)])
    })

for month, revenue in monthly_sales.items():
    summarized_data.append({
        'period': month,
        'period_type': 'Monthly',
        'total_revenue': revenue,
        'total_expenditure': sum([sale['total_expenditure'] for sale in combined_data if sale['sale_date'].startswith(str(month))]),
        'profit': sum([sale['profit'] for sale in combined_data if sale['sale_date'].startswith(str(month))])
    })

for quarter, revenue in quarterly_sales.items():
    summarized_data.append({
        'period': quarter,
        'period_type': 'Quarterly',
        'total_revenue': revenue,
        'total_expenditure': sum([sale['total_expenditure'] for sale in combined_data if f"Q{(datetime.strptime(sale['sale_date'], '%m/%d/%Y').month - 1) // 3 + 1}" == quarter]),
        'profit': sum([sale['profit'] for sale in combined_data if f"Q{(datetime.strptime(sale['sale_date'], '%m/%d/%Y').month - 1) // 3 + 1}" == quarter])
    })

# 2. Category-wise summary
for category in category_sales:
    summarized_data.append({
        'period': category,
        'period_type': 'Category',
        'total_revenue': sum([sale['total_revenue'] for sale in combined_data if sale['category'] == category]),
        'total_expenditure': sum([sale['total_expenditure'] for sale in combined_data if sale['category'] == category]),
        'profit': category_profits.get(category, 0)
    })

# Write the summarized data to the "summarized_data.csv" file
with open("summarized_data.csv", "w", newline='') as file:
    writer = csv.DictWriter(file, ['period', 'period_type', 'total_revenue', 'total_expenditure', 'profit'])
    writer.writeheader()
    writer.writerows(summarized_data)

print("Summarized report written to 'summarized_data.csv'")
```
