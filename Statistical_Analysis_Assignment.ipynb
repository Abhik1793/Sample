```python
#1.HealthCare Plus recorded the daily number of patient admissions for the past 10 days: [32, 28, 35, 30, 29, 27, 31, 34, 33, 30] 
#1. Patient Admissions – Mean, Median, Mode

import numpy as np
from scipy import stats

admissions = [32, 28, 35, 30, 29, 27, 31, 34, 33, 30]

mean = np.mean(admissions)
median = np.median(admissions)
mode = stats.mode(admissions, keepdims=False).mode

# 10% increase
increased = [x * 1.1 for x in admissions]
mean_inc = np.mean(increased)
median_inc = np.median(increased)
mode_inc = stats.mode(increased, keepdims=False).mode

print(f"Mean: {mean}, Median: {median}, Mode: {mode}")
print(f"Increased Mean: {mean_inc}, Median: {median_inc}, Mode: {mode_inc}")
```

```python
#2. The recovery duration (in days) of 10 patients who underwent the same surgery is recorded as follows: [5, 7, 6, 8, 9, 5, 6, 7, 8, 6] 
#2. Recovery Duration – Range, Variance, Standard Deviation

recovery = [5, 7, 6, 8, 9, 5, 6, 7, 8, 6]

data_range = max(recovery) - min(recovery)
variance = np.var(recovery, ddof=1)
std_dev = np.std(recovery, ddof=1)

# Add new patients
extended = recovery + [4, 10]
new_std_dev = np.std(extended, ddof=1)

print(f"Range: {data_range}, Variance: {variance}, Std Dev: {std_dev}")
print(f"New Std Dev with 2 new patients: {new_std_dev}")
```

```python
#3. Patient satisfaction scores (on a scale of 1 to 10) collected from a hospital survey are: [8, 9, 7, 8, 10, 7, 9, 6, 10, 8, 7, 9]
#3. Satisfaction Scores – Skewness & Kurtosis

satisfaction = [8, 9, 7, 8, 10, 7, 9, 6, 10, 8, 7, 9]

skewness = stats.skew(satisfaction)
kurtosis = stats.kurtosis(satisfaction)

print(f"Skewness: {skewness}, Kurtosis: {kurtosis}")
```

```python
#4. HealthCare Plus wants to analyze the relationship between nurse staffing levels and patient recovery time. Data from 6 hospital departments is provided: 
#Number of Nurses Average Recovery Time (days) Given in code:
#4. Correlation – Nurses vs Recovery Time

nurses = [10, 12, 15, 18, 20, 22]
recovery_time = [8, 7, 6, 5, 4, 3]

correlation = np.corrcoef(nurses, recovery_time)[0, 1]
print(f"Correlation Coefficient: {correlation}")
```

```python
#5. The hospital claims that the average patient wait time in the emergency department is 30 minutes. 
#A sample of 10 patient wait times (in minutes) is recorded: [32, 29, 31, 34, 33, 27, 30, 28, 35, 26] 
#Section B: Inferential Statistics & Probability

wait_times = [32, 29, 31, 34, 33, 27, 30, 28, 35, 26]

t_stat, p_value = stats.ttest_1samp(wait_times, 30)

print(f"T-Statistic: {t_stat}, P-Value: {p_value}")
```

```python
#6. A survey was conducted on hospital cleanliness and patient satisfaction. The following data was collected: 

#6. Chi-Square Test – Cleanliness vs Satisfaction

from scipy.stats import chi2_contingency

table = [
    [90, 10],
    [60, 40],
    [30, 70]
]

chi2, p, dof, expected = chi2_contingency(table)

print(f"Chi2: {chi2}, P-Value: {p}, DoF: {dof}")
```

```python
#7. The hospital tested three different treatment methods (A, B, and C) for managing post-surgery pain. 
#The recovery durations (in days) under each treatment are:
#7. ANOVA – Recovery by Treatment

#Treatment A
A = [5, 6, 7, 5, 6]
#Treatment B
B = [8, 9, 7, 8, 10]
#Treatment C
C = [4, 5, 6, 5, 4]

f_stat, p_value = stats.f_oneway(A, B, C)

print(f"F-Statistic: {f_stat}, P-Value: {p_value}")
```

```python
#The hospital administration time (in minutes) for 12 patients is recorded as: [12, 15, 14, 16, 18, 13, 14, 17, 15, 19, 16, 14] 
#8. Normality Test – Administration Time

from scipy.stats import shapiro
import matplotlib.pyplot as plt
import seaborn as sns

admin_times = [12, 15, 14, 16, 18, 13, 14, 17, 15, 19, 16, 14]

# Shapiro-Wilk test
stat, p = shapiro(admin_times)
print(f"Shapiro-Wilk Test Statistic: {stat}, P-Value: {p}")

# Histogram + QQ plot
sns.histplot(admin_times, kde=True)
plt.title("Histogram of Administration Times")
plt.show()

stats.probplot(admin_times, dist="norm", plot=plt)
plt.title("QQ Plot of Administration Times")
plt.show()
```

```python
#9. The hospital is studying the distribution of patient arrival times in the emergency department. 
#Historical data suggests that emergency cases arrive at an average rate of 5 per hour. 
#9. Poisson Distribution – Emergency Arrivals

from scipy.stats import poisson

# λ = 5 cases/hour
prob_3 = poisson.pmf(3, mu=5)

print(f"Probability of exactly 3 arrivals: {prob_3}")
```

```python
#10. The number of surgeries performed per day in the hospital follows a specific distribution pattern. 
#Historical data shows the following frequencies:
#10. Expected Surgeries – Discrete Distribution

x = np.array([0, 1, 2, 3, 4, 5])
f = np.array([5, 12, 18, 22, 15, 8])

expected_value = np.sum(x * f) / np.sum(f)
print(f"Expected number of surgeries per day: {expected_value}")
```
